### Reordering Full Original Metadata Spreadsheet by Document Group for Later Editing
To be edited later inside the uiuc-ischool org repostiry

```python
import pandas as pd
```

```python
# Building Data

# Adjust filenames as needed
rels = pd.read_csv("/Users/owenmonroe/Desktop/GitHub/lonewoman_cb_test/full_metadata_reordering_work/experiment_data/full_schwebel_document_groups.csv")  # your relationships doc
full = pd.read_csv("/Users/owenmonroe/Desktop/GitHub/lonewoman_cb_test/full_metadata_reordering_work/experiment_data/full_schwebel_metadata.csv")  # your full metadata sheet

# Make sure these names match your actual column headers exactly:
col_orig = "Document Key of Original"
col_rep  = "Document Key of Reprint/Related"

order = []
current_original = None

for _, row in rels.iterrows():
    orig = row[col_orig]
    rep  = row[col_rep]

    # If there is a new Original, remember it and add it once
    if isinstance(orig, str) and orig.strip():
        current_original = orig.strip()
        order.append(current_original)

    # If there is a Reprint/Related, append it after the current original
    if isinstance(rep, str) and rep.strip():
        order.append(rep.strip())

# Optional: check what we got
print("Order from relationships sheet:")
print(order)
```

```python
# Mappign to resort

full = full.rename(columns={col_orig: "id"})

# Create a dictionary mapping item → order index
order_map = {obj_id: i for i, obj_id in enumerate(order)}

# Create a sorting key column
full["sort_key"] = full["id"].map(order_map)

# Sort: items not in relationships will go LAST
full_sorted = full.sort_values("sort_key", na_position="last")

# Diagnostics
missing_in_full = [obj for obj in order if obj not in full["id"].values]
if missing_in_full:
    print("IDs in relationships but NOT found in full metadata:", missing_in_full)

extra_in_full = [obj for obj in full["id"].values if obj not in order]
if extra_in_full:
    print("IDs in full metadata but NOT found in relationships:", extra_in_full)

# Rename back to original column name
full_sorted = full_sorted.rename(columns={"id": col_orig})

# Remove helper column
full_sorted = full_sorted.drop(columns=["sort_key"])

# Save final sorted file
full_sorted.to_csv("full_metadata_sorted.csv", index=False)

print("\n✔ Sorting complete! Saved as full_metadata_sorted.csv")
```

```python
# Cleaning Data

sorted_df = pd.read_csv("/Users/owenmonroe/Desktop/GitHub/lonewoman_cb_test/full_metadata_reordering_work/experiment_data/full_metadata_sorted.csv")

# 2. Select all text (object) columns to clean
text_cols = sorted_df.select_dtypes(include="object").columns

for col in text_cols:
    # Skip entirely empty columns if you like (optional)
    if sorted_df[col].isna().all():
        continue

    # a) Remove <i> and </i> tags
    sorted_df[col] = sorted_df[col].str.replace("<i>", "", regex=False)
    sorted_df[col] = sorted_df[col].str.replace("</i>", "", regex=False)

    # b) Remove commas immediately before straight quotes: ," 
    sorted_df[col] = sorted_df[col].str.replace(r',(?=")', '', regex=True)

    # c) Remove commas immediately before curly closing quotes: ,”
    sorted_df[col] = sorted_df[col].str.replace(r',(?=”)', '', regex=True)

    # NEW: Remove trailing commas at end of entire cell
    sorted_df[col] = sorted_df[col].str.replace(r',$', '', regex=True)

    # d) (Optional) trim extra spaces
    sorted_df[col] = sorted_df[col].str.strip()

# 3. Save cleaned metadata
sorted_df.to_csv("/Users/owenmonroe/Desktop/GitHub/lonewoman_cb_test/full_metadata_reordering_work/experiment_data/full_metadata_cleaned.csv", index=False)

print("✔ Cleaned file written to full_metadata_cleaned.csv")


```
